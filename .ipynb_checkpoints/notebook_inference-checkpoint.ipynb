```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df_clean = pd.read_csv ('dataframeclean.csv')
```

```python
# Household too correlated so we don't use it as a features
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(df_clean.corr(),annot=True)
```

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(df_clean.corr()[['households']].sort_values(by='households', ascending=False), vmin=-1, vmax=1, annot=True, cmap='flare')
heatmap.set_title('Features Correlating with households', fontdict={'fontsize':18}, pad=16);
```
